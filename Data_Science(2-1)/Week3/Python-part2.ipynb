```python
from time import time
start_time = time()
x = [4, None, 'foo']
x.extend([7, 8, (2, 3)]) # x의 객체를 업데이트
print(x)
end_time = time()
print("Time elapsed: ", end_time - start_time) # 더 빠름
```

```python
import bisect
c = [1, 2, 2, 2, 3, 4, 7]
print(bisect.bisect(c, 2)) # 4 (2가 4번째로 들어가야 함)
print(bisect.bisect(c, 5)) # 6 (5가 6번째로 들어가야 함)
print(bisect.insort(c, 6)) # None (c에 6을 추가하고 None을 반환) inplace
print(c)
```

```python
import re
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip() # stripping whitespace
        value = re.sub('[!#?]', '', value) # removing punctuation symbols
        value = value.title() # standardizing on proper capitalization
        result.append(value)
    return result

states = [' Alabama ', 'Georgia!', 'georgia', 'West virginia?']
clean_strings(states)
```

```python
# 위의 코드를 map 함수를 사용하여 간단하게 작성(함수도 객체이기 때문에 인자로 넘길 수 있음)
states = [' Alabama ', 'Georgia!', 'georgia', 'West virginia?']
def remove_punctuation(value):
    return re.sub('[!#?]', '', value)
clean_ops = [str.strip, remove_punctuation, str.title]
def clean_strings(strings, ops):
    result = []
    for value in strings:
        for function in ops:
            value = function(value)
        result.append(value)
    return result
clean_strings(states, clean_ops)
```

```python
def apply_to_list(some_list, f):
    return [f(x) for x in some_list]
def mul2(x):
    return x * 2
ints = [4, 0, 1, 5, 6]
apply_to_list(ints, lambda x: x * 2)
```

```python
# 제너레이터
# 리스트 컴프리헨션과 유사하지만, 대괄호 대신 괄호를 사용
# 값을 미리 만들어두지 않고, 필요할 때 값을 만들어내는 방식
# 제너레이터 표현식은 한 번에 한 요소를 만들어냄(메모리를 더 효율적으로 사용) / 리스트 컴프리헨션은 리스트 전체를 만들어냄
gen = (x * 2 for x in range(100))
print(gen)
for i in gen:
    print(i)
    if i > 10:
        break
```
