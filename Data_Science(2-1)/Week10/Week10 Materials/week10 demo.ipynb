# Histogram and kde plot

```python
df = pd.read_csv('Titanic.csv')
df
```

```python
df['Age'].hist(bins=20) # draw the histogram with 20 bins
```

```python
df['Age'].plot.kde() # this will show some false data less than 0, and over the max age
```

```python
sns.kdeplot(df['Age'], cut=0, bw_adjust=1)
plt.xlim(df['Age'].min(), df['Age'].max())  # trim the plot with the min/max of the Age
plt.show()
```
