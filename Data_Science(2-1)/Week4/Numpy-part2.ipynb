```python
import numpy as np
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True,False,True,True,False])
result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)] 
print(result, "\n")
result = np.where(cond, xarr, yarr) # 조건이 참이면 xarr, 거짓이면 yarr
print(result)
```

```python
arr = np.random.randn(4, 4)
print(arr)
result = np.where(arr > 0, 2, -2)
print(result, "\n")
math = {'sum': result.sum(), 'mean': result.mean(), 'std': result.std(), 'var': result.var(), 'min': result.min(), 'max': result.max(), 'argmin': result.argmin(), 'argmax': result.argmax(), 'cumsum': result.cumsum(), 'cumprod': result.cumprod()}
print(math, "\n")
print(result.sum(axis=0), "\n") # 열의 합
print((result > 0).sum()) # 양수인 원소의 개수
print((result > 0).any()) # 하나라도 참이 있는지
print((result > 0).all()) # 모두 참인지
```

```python
print(np.sort(result[:2,:], axis=1)) # 열로 훑어서 정렬
```

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
print(np.unique(names)) # 중복 제거
```

```python
values = np.array([6, 0, 0, 3, 2, 5, 6])
print(np.in1d(values, [2, 3, 6, 7])) # values에 2, 3, 6이 있는지
print(np.intersect1d(values, [2, 3, 6, 7])) # values와 [2, 3, 6, 7]의 교집합
print(np.union1d(values, [2, 3, 6, 7])) # values와 [2, 3, 6, 7]의 합집합
print(np.setdiff1d(values, [2, 3, 6, 7])) # values에만 있는 원소
print(np.setxor1d(values, [2, 3, 6, 7])) # 한 배열에만 있는 원소
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])
print(np.diag(A)) # 대각선 원소
print(np.dot(A, B)) # 행렬 곱셈
print(np.trace(A)) # 대각선 원소의 합
print(np.linalg.det(A[:2,:2])) # 행렬식
print(np.linalg.inv(A[:2,:2])) # 역행렬
print(np.linalg.eig(A[:2,:2])) # 고유값, 고유벡터
print(np.linalg.qr(A)) # QR 분해
print(np.linalg.svd(A)) # 특이값 분해
print(np.linalg.solve(A[:2,:2], [6, 14])) # Ax = b의 해
print(np.linalg.lstsq(A, [6, 14])) # 최소자승 문제의 해
```
