````python
import re
txt = "The rain in Spain"
x = re.search("ai", txt) # Return a Match object
type(x)
re.Match
print("{0} {1}".format(x.start(),x.end())) # the start-, and end positions
print(txt[x.start():x.end()])
````

```Python
In [20]: x.group() # returns the part of the string where there was a match
Out[20]: ai
In [21]: x.span() # returns a tuple containing the start-, and end positions
Out[21]: (5, 7)
In [22]: x.string # returns the string passed into the function
Out[22]: The rain in Spain
```

````python
txt = "The rain in Spain"
x = re.findall("ai", txt) # Returns a list containing all matches
print(type(x)) 
print(x) 
len(x)
````
