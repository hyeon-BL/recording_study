Selection sort 알고리즘

```python
A = [30, 412, 21, 404, 42]
n = len(A)
def selection_sort(lst, n):
    for i in range(n):
        min_index = i
        for j in range(i+1, n):               # 가장 작은 원소를 찾아서 맨 앞으로 보내는 과정
            if lst[j] < lst[min_index]:
                min_index = j
        lst[i], lst[min_index] = lst[min_index], lst[i]
    return lst
print(selection_sort(A, n))
```

알고리즘 분석<br>
i = 1일 때 4번 반복<br>
i = 2일 때 3번 반복<br>
i = 3일 때 2번 반복<br>
i = 4일 때 1번 반복<br>
<br>
-> 총 n(n-1)/2번 반복<br>
-> 대략 n²으로 시간 증가

좌측 숫자는 컴퓨터의 메모리에 접근하는 횟수를 나타냄<br>
 : 4n+5 <= 작동시간(횟수) <= 5n+5
