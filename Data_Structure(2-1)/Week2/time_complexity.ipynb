Selection sort 

```python
A = [30, 412, 21, 404, 42]
n = len(A)
def selection_sort(lst, n):
    for i in range(n):
        min_index = i
        for j in range(i+1, n):               #        
            if lst[j] < lst[min_index]:
                min_index = j
        lst[i], lst[min_index] = lst[min_index], lst[i]
    return lst
print(selection_sort(A, n))
```
