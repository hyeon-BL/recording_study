### Selection sort 알고리즘

```python
A = [30, 412, 21, 404, 42]
n = len(A)
def selection_sort(lst, n):
    for i in range(n):
        min_index = i
        for j in range(i+1, n):               # 가장 작은 원소를 찾아서 맨 앞으로 보내는 과정
            if lst[j] < lst[min_index]:
                min_index = j
        lst[i], lst[min_index] = lst[min_index], lst[i]
    return lst
print(selection_sort(A, n))
```

### Big O notation


```python
def selection_sort(lst, n): # 위 코드 참조                 1
    for i in range(n):                                   #n
        min_index = i                                    #n 
        for j in range(i+1, n):                          #n^2
            if lst[j] < lst[min_index]:                  #n^2
                min_index = j                            #0~n^2
        lst[i], lst[min_index] = lst[min_index], lst[i]  #n
    return lst                                           #1
```

2n²+3n+1 ~ 3n²+3n+1<br>
worst case 기준 : 3n²+3n+1 = f(n)
g(n) = O(n²) -> n² class라는 것이 중요
- Operation의 가장 작은 지수
- Operation의 상수항은 제거
